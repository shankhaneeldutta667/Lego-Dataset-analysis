```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('lego_sets.csv')
df1.head()
```

```python
df2=pd.read_csv('parent_themes.csv')
df2.head()
```

### Observation:  key 'parent_theme' in lego_sets.csv file matches with key  'name' in parent_themes.csv
- We can use pandas merge() to merge two dataframes and join them based on the common keys !
    - If both key columns contain rows where the key is a null value, those rows will be matched against each other.                   This is different from usual SQL join behaviour and can lead to unexpected results.

```python
merged_df=df1.merge(df2,left_on='parent_theme',right_on='name')
merged_df.head(10)
```

### Observations: 
- While merging two DataFrames into a single Dataframe, **IF there is a(or many) columns that have SAME name** in both Dataframes,
  then pandas will AUTOMATICALLY give x,y or any **special notation to those columns**.
  Eg: **name_x from df1 and name_y from df2**

### Observations:
- In the single Dataframe, we can see that column **name_y** is duplicate of column **parent_theme**
    - We'll **drop name_y** column, and work with only **parent_theme** column using drop(columns='column name',inplace=True)

```python
merged_df
```

```python
new_df=merged_df.drop(columns='name_y')
new_df
```

```python
new_df.head()
```

## Filter out the licensed themes

### Q2) In which year Star Wars wasn't the popular licensed theme(in terms of number of sets released that year) ?

```python
licensed.head(10)
```

- Group the dataframe based on 'year' and 'parent_theme'
    
    - obeserve the count of themes each year

```python
licensed['count']=1
grouped_df=licensed.groupby(by=['year','parent_theme']).sum().reset_index()
```

```python
grouped_df
```

```python
max_df=grouped_df.sort_values(['count'],ascending=False).drop_duplicates(['year'])
```

```python
max_df.sort_values(by=['year'],inplace=True)
max_df
```

```python
# we can clearly analyze that in year 2017, Starwars got beat out because of 'Super Heroes' having count 72
```
